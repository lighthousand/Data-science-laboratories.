# Reprezentarea grafia a datelor

## 1. Histograma

```python
import pandas as pd
print("", pd.__version__)
url_balance = 'http://archive.ics.uci.edu/ml/machine-learning-databases/balance-scale/balance-scale.data'
names = ['class_name', 'left_weight', 'left_distance', 'right_weight', 'right_distance']
balance_df = pd.read_csv(url_balance, names=names)
balance_df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
balance_df.hist(by=balance_df['right_weight'])
```

## 2. Boxplots

```python
balance_df.boxplot()
```

## 3. Scatter plots

```python
balance_df.plot.scatter(1,1)
```

## 4. Matrix scatter plots

```python
balance_df.plot()
```

## 5. Vizualizare de matrice de corelatie

```python
plt.matshow(balance_df.corr())
```

## 6. Coordonate paralele 

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(balance_df,"class_name")
plt.show()
```

## 7. Alte tipuri de grafice

```python
from pandas.plotting import andrews_curves

andrews_curves(balance_df, 'class_name')
```

```python
from pandas.plotting import lag_plot

lag_plot(balance_df)
```

```python
from pandas.plotting import radviz

plt.figure()

radviz(balance_df, 'class_name')
```
